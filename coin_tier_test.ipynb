```python
coin_tier_df = pd.read_csv('./tier_data/dataframe/all.csv')
coin_tier_df['tier'] = coin_tier_df['tier'].map(int)
coin_tier_df[ (coin_tier_df['date']=='2021-07-28') & (coin_tier_df['tier']<2) ].alias
```

```python
coins = list(set(coin_tier_df[ (coin_tier_df['date']==date) & (coin_tier_df['tier']<2) ].alias))
coins
```

```python
from tqdm import tqdm

df = {}
for coin in tqdm(coins):
    df[coin] = getCoinInfo.getCoinDF('KRW-'+coin,30,200,size=24*2*7)
  
df[coins[1]]
```

```python
today = '2021-07-29T00:00:00'
before = '2021-07-28T00:00:00'
```

```python
for coin in coins:
    today_price = df[coin][df[coin]['candle_date_time_kst']==today]['opening_price']
    before_price = df[coin][df[coin]['candle_date_time_kst']==before]['opening_price']
    print(today_price.iloc[0] - before_price.iloc[0])
```
